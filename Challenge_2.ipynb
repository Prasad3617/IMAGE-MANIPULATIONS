### Defining a function to perform gaussian blur on the same image taken as input.

#### *We provide details as the filter size and the value for standard deviation as input.*

```python
def gaussian_filter(image,filter_size,sigma):
    filtered_image=cv2.GaussianBlur(image,(filter_size,filter_size),sigma)
    return filtered_image
```

### Defining a function to perform PREWITT FILTER OPERATIONS on the same image taken as input.

#### *We  the mathematical values for the matrix as mask. "Horizontal" and "Vertical Filter(s)" are used in the code.*
#### *We also perform mathematical operations of taking square root and taking absolute value: "abs_grad_x/abs_grad_y"*

```python
def prewitt_filter(image):
    image = gaussian_filter(image,5,5)
    vertical_filter = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
    horizontal_filter = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
    vertical_filtered = cv2.filter2D(image,-1,vertical_filter)
    horizontal_filtered = cv2.filter2D(image,-1,horizontal_filter)
    abs_grad_x = cv2.convertScaleAbs(vertical_filtered)
    abs_grad_y = cv2.convertScaleAbs(horizontal_filtered)
    grad = cv2.addWeighted(abs_grad_x,0.5,abs_grad_y,0.5,0)
    return grad
```

```python
image = cv2.imread('/content/drive/MyDrive/Challenge_2.png')
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
prewitt_edges = prewitt_filter(gray_image)
cv2_imshow(prewitt_edges)
cv2.waitKey(0)
```

### Defining the basic properties for the "Laplacian - Filter" in the image as output using the logic provide below.

#### *Laplacian filter: dst = cv.Laplacian(src_gray, ddepth, ksize=kernel_size)*

```python
image = cv2.imread('/content/drive/MyDrive/Challenge_2.png')
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
laplacian_filter = cv2.Laplacian(gray_image,16, 3)
abs_laplace_filter = cv2.convertScaleAbs(laplacian_filter)
cv2_imshow(laplacian_filter)
cv2.waitKey(0)
```
