```python
import cv2
import numpy as np
```

```python
def sp_noise(image):
    output = image.copy()
    if len(image.shape) == 2:
        black = 0
        white = 255
    else:
        black = np.array([0,0,0],dtype='uint8')
        white = np.array([255,255,255],dtype='uint8')
        random_threshold = np.random.random(image.shape[:2])
        output[random_threshold<(random_threshold/2)] = black
        output[random_threshold>1-(random_threshold/2)] = white
        return output
```

```python
def median_filter(image,filter_size):
    median_filtered_image = cv2.medianBlur(image,filter_size)
    return median_filtered_image
```

```python
image = cv2.imread('desktop/untitled folder 3/a.jpg')
sp_noisy_image = sp_noise(image)
median_filtered_image = median_filter(sp_noisy_image,3)
cv2.imshow('median filtered image',np.hstack((sp_noisy_image,median_filtered_image)))
cv2.waitKey(0)
```
